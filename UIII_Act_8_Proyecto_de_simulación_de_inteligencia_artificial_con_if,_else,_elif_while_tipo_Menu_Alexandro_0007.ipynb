### Zapatero y Zapato

```python
ejecutando = True
tipo_zapato = "No registrado"
material = "Desconocido"
nivel_desgaste = 0

print("--- SISTEMA IA: EL ASISTENTE DEL MAESTRO ZAPATERO ---")

while ejecutando == True:
    print("\n--- MENÚ DE TALLER ---")
    print("1. Registrar nuevo calzado")
    print("2. Análisis de reparación (Lógica IA)")
    print("3. Ver tabla de conocimientos de la IA")
    print("4. Apagar sistema")
    opcion = input("IA: Seleccione una opción (1-4): ")

    # CASO 1: Registro del calzado
    if opcion == "1":
        tipo_zapato = input("Ingrese el tipo de zapato (ej. Bota, Tenis, Zapatilla): ")
        material = input("Ingrese el material (ej. Cuero, Tela, Sintetico): ")
        print("IA: Calzado registrado: " + tipo_zapato + " de " + material + ".")

    # CASO 2: Simulación de Inteligencia Artificial (Diagnóstico) [4.6, 18.8.6]
    elif opcion == "2":
        print("IA: Iniciando escaneo de daños...")
        nivel_desgaste = int(input("Ingrese nivel de desgaste de la suela (1-10): "))

        # Lógica de decisión con operadores lógicos y relacionales [3.7.3]
        if material == "Cuero" and nivel_desgaste <= 5:
            print("IA DIAGNÓSTICO: El material es resistente y el daño es superficial.")
            print("IA ACCIÓN: Se recomienda limpieza profunda y boleado con cera.")

        elif material == "Tela" or tipo_zapato == "Tenis":
            if nivel_desgaste > 7:
                print("IA DIAGNÓSTICO: Estructura de tela muy dañada.")
                print("IA ACCIÓN: No es rentable reparar. Sugerir cambio de calzado.")
            else:
                print("IA DIAGNÓSTICO: Desgaste moderado.")
                print("IA ACCIÓN: Reparación mediante pegado térmico y lavado.")

        else:
            print("IA ESTADO: Caso complejo. Requiere revisión manual del Maestro Zapatero.")

    # CASO 3: Tabla de lógica integrada en el código [18.10.1]
    elif opcion == "3":
        print("\n--- TABLA DE LÓGICA IA: REPARACIONES ---")
        print("MATERIAL | DESGASTE | RESULTADO")
        print("-" * 45)
        print("Cuero    | 1 a 5    | Limpieza y Cera")
        print("Tela     | > 7      | Sugerir Nuevo")
        print("Tela     | 1 a 7    | Pegado Térmico")
        print("Otro     | Cualquiera| Revisión Manual")
        print("-" * 45)

    # CASO 4: Salida del sistema [5.2]
    elif opcion == "4":
        print("IA: Cerrando taller... ¡Hasta mañana, Maestro Zapatero!")
        ejecutando = False
print("Elaborado por Alexandro Aguilar NL = 1")
```
